### Importing essential libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading the data and creating the DataFrame

```python
# Creating a dictionary of required links 
urls = {'GDP': "https://clck.ru/32VMek",
        'PCEC': "https://clck.ru/32VMfq",
        'Investments': "https://clck.ru/32VMgX",
        'Inventories': "https://clck.ru/32VMh8",
        'GDP_Implicit_Price_Deflator_inflation': "https://clck.ru/32VMdp",
        'CPI_less_food_and_energy': "https://clck.ru/32VMdL",
        'total_CPI': "https://clck.ru/32VMb6",
        'Wilshire_5000_Full_Cap_Price_Index': "https://clck.ru/32VMZr",
        'Real_Estate_Prices': "https://clck.ru/32VLkN",
        'Short-term_rate_nominal': "https://clck.ru/32VK7V",
        'Long-term_rate_nominal': "https://clck.ru/32VJsY",
        'Yield_curve': "https://clck.ru/32VKKR",
        'Real_ef_exchange_rate': "https://clck.ru/32VM8X",
        'DM_USD_exchange_rate': "https://clck.ru/32VMS5",
        'M1':"https://clck.ru/32VP7t",
        'M3': "https://clck.ru/32VPDj",
        'Total_loans': "https://clck.ru/32VTpu"}

# Creating and populating the DataFrame with data (dates and values for each required time series)
df = pd.DataFrame()
for key in urls:
    df[f"{key} Date"] = pd.read_csv(urls[key]).iloc[:,0]
    df[key] = pd.read_csv(urls[key]).iloc[:,1]
    
# Creating and populating a dictionary by assigning to key (time series name) its DataFrame (dates and vaules)
components = {key: None for key in list(urls.keys())}
for i in range(len(urls.keys())):
    components[list(urls.keys())[i]] = pd.DataFrame(df.iloc[:,2*i]).join(pd.DataFrame(df.iloc[:,2*i+1]))
    
# Creating the final DataFrame
data_components = pd.DataFrame({'GDP Date':components['GDP'].iloc[:,0],'GDP':components['GDP'].iloc[:,1]}).rename(columns={'GDP Date':'Date'})
for i in range(len(list(components.keys()))-1):
    name = list(components.keys())[i+1]
    data_components = data_components.merge(components[list(components.keys())[i+1]],
                                           how='left', left_on='Date', right_on=f"{name} Date").drop(columns=f"{name} Date")

data_components = data_components.set_index('Date', drop=True)
data_components.loc[data_components['Real_Estate_Prices'] == '.', 'Real_Estate_Prices'] = ''
data_components.loc[data_components['Long-term_rate_nominal'] == '.', 'Long-term_rate_nominal'] = ''
```
